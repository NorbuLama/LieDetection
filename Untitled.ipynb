```python
def save_mfccs(json_path):
    data = {
            "type"  : [],
            "label" : [],
            "mfccs" : []}

    data["type"].append("lie")
    data["type"].append("truth")

    
    for i in audiopath1:
        signal, sr = librosa.load(i, sr =22050)
        mfcc = librosa.feature.mfcc(signal, sr=sr, n_fft = 2048, n_mfcc=20, hop_length = 512)
        mfcc= mfcc.T

        data["label"].append(1)
        data["mfccs"].append(mfcc.tolist())
        



    for i in audiopath2:
        signal, sr = librosa.load(i, sr =22050)
        signal, sr = librosa.load(i, sr =22050)
        mfcc = librosa.feature.mfcc(signal, sr=sr, n_fft = 2048, n_mfcc=20, hop_length = 512)
        mfcc= mfcc.T

        data["label"].append(0)
        data["mfccs"].append(mfcc.tolist())
        

    with open(json_path, "w") as fp:
        json.dump(data,fp, indent =4)
            
```

```python
json_path = json_path = "/Users/norbulama/Desktop/CS-4000/data.json"
if __name__ == "__main__":
    save_mfccs(json_path)
```
